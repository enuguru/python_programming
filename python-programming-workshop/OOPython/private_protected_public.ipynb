```python
# program to show public access modifier in a class
 
# super class
class Student:
    
     # constructor
     def __init__(self, name, roll, branch): 
          self.name = name
          self.roll = roll
          self.branch = branch
          self.section = 2
    
     # public member function  
     def displayRollAndBranch(self):
          # accessing protected data members
          print("Roll: ", self.roll)
          print("Branch: ", self.branch)


# derived class
class Geek(Student):
 
       # constructor
       def __init__(self, name, roll, branch):
                Student.__init__(self, name, roll, branch)
         
       # public member function
       def displayDetails(self):
                   
                 # accessing protected data members of super class
                print("Name: ", self.name)
                   
                 # accessing protected member functions of super class
                self.displayRollAndBranch()
                print("section: ", self.section)


# this is the main program part
# creating objects of the derived class       
obj = Geek("R2J", 1706256, "Information Technology")
print("Name: ", obj.name)
# calling public member functions of the class
obj.displayDetails()
obj.displayRollAndBranch()
```

```python
# program to illustrate protected access modifier in a class

# super class
class Student:
    
     # constructor
     def __init__(self, name, roll, branch): 
          self._name = name
          self._roll = roll
          self._branch = branch
          self._section = 2
    
     # public member function
     def _displayRollAndBranch(self):
          # accessing protected data members
          print("Roll: ", self._roll)
          print("Branch: ", self._branch)
            
     # public member function  
     def displayRollAndBranch(self):
          # accessing protected data members
          print("Roll: ", self._roll)
          print("Branch: ", self._branch)


# derived class
class Geek(Student):
 
       # constructor
       def __init__(self, name, roll, branch):
                Student.__init__(self, name, roll, branch)
         
       # public member function
       def displayDetails(self):
                   
                 # accessing protected data members of super class
                print("Name: ", self._name)
                   
                 # accessing protected member functions of super class
                self._displayRollAndBranch()
                print("section: ", self._section)


# this is the main program part
# creating objects of the derived class       
obj = Geek("R2J", 1706256, "Information Technology")
print("Name: ", obj._name)
# calling public member functions of the class
obj.displayDetails()
obj._displayRollAndBranch()
```
